## Data wrangling


```python
df_anime = pd.read_csv('/content/anime.csv')
df_rating = pd.read_csv('/content/rating.csv')
```

Vamos unir os dois dataframe em um só pela coluna 'user_id' e excluir a coluna 'rating' no dataset df_anime. Pois o que nos interessa é só o rating (avaliação) do usuário.

```python
df= pd.merge(df_rating, df_anime.drop('rating',axis=1),on='anime_id')
print(df.shape)
df.head()
```

Aqui podemos perceber que o  mesmo anime aparece várias vezes, então temos que agrupar esse "rating" para o número de avaliações que cada anime recebeu.


```python
df.groupby('name')['rating'].count().sort_values(ascending=False).head(600)
```

Criar um novo dataframe cpm o número de avaliações e a média de avaliações

```python
new_df = pd.DataFrame(df.groupby('name')['rating'].mean())
new_df['Quantidade'] = pd.DataFrame(df.groupby('name')['rating'].count())
new_df['avaliação média'] = pd.DataFrame(df.groupby('name')['rating'].mean().round(2))
new_df
```

Criar um novo dataframe com nome e genero e utilizar o nome do anime como index no dataframe


```python
df_genero = pd.DataFrame(data=df_anime[['name','genre']])
df_genero.set_index('name',inplace=True)
df_genero
```

## Criação do Sistema de Recomendação


```python
def verificar_genero(lista_genero, string): #Definir a função, as variaveis são a lista de genero e uma string
  if any(x in string for x in lista_genero): #Se existir um x nessa string que está dentro da lista de generos
   return True #Vai retornar True
  else: #Se não
   return False #Vai retornar False 
```

```python
genero_anime = df_genero.loc['Death Note'].values[0].split(', ')
genero_anime
```

```python
cols = df_anime[df_anime['genre'].apply(lambda x: verificar_genero(genero_anime, str(x)))]['name'].tolist()
```

```python
matriz_de_animes = df[df['name'].isin(cols)].pivot_table(index='user_id', columns='name', values='rating')
```

```python
matriz_de_animes.head()
```

```python
anime_nota = matriz_de_animes['Death Note']
anime_nota
```

```python
anime_parecido = matriz_de_animes.corrwith(anime_nota)
```

```python
anime_correlacionado = pd.DataFrame(anime_parecido, columns=['correlação'])
anime_correlacionado.head()
```

```python
anime_correlacionado = anime_correlacionado.join(new_df[['Quantidade', 'avaliação média']])
```

```python
anime_correlacionado.head()
```

```python
anime_correlacionado.dropna(inplace=True)
```

```python
animes_recomendados = anime_correlacionado[anime_correlacionado['Quantidade']> 4000].sort_values('correlação', ascending = False)
```

```python
animes_recomendados = animes_recomendados.rename_axis('Animes Recomendados')
```

## Modelo

```python
def recomendar_anime(nome_do_anime, n):
  genero_anime = df_genero.loc[nome_do_anime].values[0].split(', ')
  cols = df_anime[df_anime['genre'].apply(lambda x: verificar_genero(genero_anime, str(x)))]['name'].tolist()
  matriz_de_animes = df[df['name'].isin(cols)].pivot_table(index='user_id', columns='name', values='rating')
  anime_nota = matriz_de_animes[nome_do_anime]
  anime_parecido = matriz_de_animes.corrwith(anime_nota)
  anime_correlacionado = pd.DataFrame(anime_parecido, columns=['correlação'])
  anime_correlacionado = anime_correlacionado.join(new_df[['Quantidade', 'avaliação média']])
  anime_correlacionado.dropna(inplace=True)
  animes_recomendados = anime_correlacionado[anime_correlacionado['Quantidade'] > 4000].sort_values('correlação', ascending=False)
#Criar nosso dataframe final, que vai ser a tabela dos animes recomendados
  animes_recomendados= animes_recomendados.rename_axis('Animes recomendados')
  print(f'Anime escolhido: {nome_do_anime}')
  return animes_recomendados.head(n+1)
```

## Recomendador


```python
recomendar_anime('Dragon Ball Z', 5)
```
